Creating subplots

```python
# the data
names = ['A', 'B', 'C']
values = [5, 10, 30]
# total figure size
plt.figure(figsize=(15,3))

# Defining subplots. We will have a matrix 1x4 (1 row and 4 columns)
plt.subplot(141)
plt.bar(names, values, alpha=0.8)
plt.axis([-1,3,0,31])

plt.subplot(142)
plt.scatter(names, values)
plt.axis([-1,3,0,31])

plt.subplot(143)
plt.plot(names, values)
plt.axis([-1,3,0,31])

plt.subplot(144)
plt.pie(values, labels=names)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
names = ['A', 'B', 'C']
values = [5, 10, 30]
# total figure size 
plt.figure(figsize=(8,8))

# Defining subplots. We will have a matrix 2x2 (2 rows and 2 columns)
plt.subplot(221)
plt.bar(names, values, alpha=0.8)
plt.axis([-1,3,0,31])

plt.subplot(222)
plt.scatter(names, values)
plt.axis([-1,3,0,31])

plt.subplot(223)
plt.plot(names, values)
plt.axis([-1,3,0,31])

plt.subplot(224)
plt.pie(values, labels=names)

plt.suptitle('Categorical Plotting')
plt.show()
```

Adding text

`xlabel`, `ylabel` and `title` are used to add text in the indicated locations. Besides, you can add text in an arbitrary location.

```python
# Generating data
mu, sigma = 100, 10
x = np.random.normal(mu, sigma, 10000)  # x is normally distributed with mean mu and standard deviation sigma
```

```python
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='r', alpha=0.6)
plt.xlabel('x values')
plt.ylabel('Probability')
plt.title('Histogram')
plt.text(70, .035, r'$\mu=100,\ \sigma=10$')
plt.axis([60,140,0,0.045])
plt.show()
```

```python
# Modifying: 
# facecolor = 'g' (green), alpha = 0.3 (transparency of the graph)
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.3)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Probability',fontweight='bold')
# left title 
plt.title('Histogram',loc='left',fontweight='bold')
plt.text(70, .035, r'$\mu=100,\ \sigma=10$')
plt.axis([60,140,0,0.045])
plt.show()
```

```python
# Modifying: 
n, bins, patches = plt.hist(x, 50, density=1, facecolor='b', alpha=0.2)
plt.xlabel('x values', fontdict={'size':11, 'weight':'regular', 'color':'gray'})
plt.ylabel('Probability', fontdict={'size':11, 'weight':'bold', 'color':'blue'})
plt.title('Histogram', loc='center', fontdict={'family':'serif','size':20, 'weight':'normal','color':'darkblue'})
plt.text(70, .035, r'$\mu=100,\ \sigma=10$',fontdict={'size':12, 'weight':'bold', 'color':'darkred'})
plt.axis([60,140,0,0.045])
plt.show()
```

```python
x = np.arange(0.0, 6.0, 0.01)
y = np.sin(2*np.pi*x)
line, = plt.plot(x, y)

# Adding annotations
plt.annotate('local max', xy=(2.3, 1), xytext=(3, 1.5),arrowprops=dict(facecolor='darkgreen'),)
plt.annotate('local min', xy=(2.8, -1), xytext=(3.5, -1.5),arrowprops=dict(facecolor='darkred'),)

plt.ylim(-2, 2)
plt.show()
```
