## Creating subplots

To create subplots you must use `plt.subplot()`

`plt.subplot(nrows, ncols, index)` 

`nrows` and `ncols` stand for the number of rows and number of columns respectively. 

If you want a 2×2 grid, set nrows=2 and ncols=2. 

For a 3×1 grid, it’s nrows=3 and ncols=1. 

The index is the subplot you want to select.

```python
# the data
names = ['A', 'B', 'C']
values = [5, 10, 30]
# total figure size
plt.figure(figsize=(15,3))

# Defining subplots. We will have a matrix 1x4 (1 row and 4 columns)
plt.subplot(1,4,1)
plt.bar(names, values, alpha=0.8)
plt.axis([-1,3,0,31])

plt.subplot(1,4,2)
plt.scatter(names, values)
plt.axis([-1,3,0,31])

plt.subplot(1,4,3)
plt.plot(names, values)
plt.axis([-1,3,0,31])

plt.subplot(1,4,4)
plt.pie(values, labels=names)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
names = ['A', 'B', 'C']
values = [5, 10, 30]
# total figure size 
plt.figure(figsize=(8,8))

# Defining subplots. We will have a matrix 2x2 (2 rows and 2 columns)
plt.subplot(221)
plt.bar(names, values, alpha=0.8)
plt.axis([-1,3,0,31])

plt.subplot(222)
plt.scatter(names, values)
plt.axis([-1,3,0,31])

plt.subplot(223)
plt.plot(names, values)
plt.axis([-1,3,0,31])

plt.subplot(224)
plt.pie(values, labels=names)

plt.suptitle('Categorical Plotting')
plt.show()
```

## Histograms

Generating data:

 x is normally distributed with mean mu and standard deviation sigma

```python
mu, sigma = 100, 10
x = np.random.normal(mu, sigma, 10000) 
```

The histogram of the data:

```python
plt.hist(x, facecolor='r', alpha=0.6);
```

### Using the `bins` parameter

The "bin" in a histogram is the interval or spacing on the x-axis. A histogram counts the number of observations that fall into each of the disjoint categories (known as bins).

```python
# Including bins=20
plt.hist(x, bins=20, facecolor='r', alpha=0.6);
```

```python
# Including bins=50
plt.hist(x, bins=50, facecolor='r', alpha=0.6);
```

### Adding labels

`xlabel`, `ylabel` and `title` are used to add text in the indicated locations. Besides, you can add text in an arbitrary location.

```python
# Adding x and y labels
plt.hist(x, bins=50, facecolor='r', alpha=0.6)
plt.xlabel('x values')
plt.ylabel('Frequency');
```

```python
# Modifying labels atributes
plt.hist(x, bins=50, facecolor='r', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold');
```

```python
# Adding a title
plt.hist(x, bins=50, facecolor='r', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.title('Histogram');
```

```python
# Modifying title 
plt.hist(x, bins=50, facecolor='r', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.title('Histogram', loc='center', fontdict={'family':'serif','size':20,\
    'weight':'normal','color':'black'});
```

### Adding text

You can add text to Matplotlib graph with `plt.text`. You have to specify the coordinates. 

```python
# Adding a text:
plt.hist(x, bins=50, facecolor='r', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.title('Histogram', loc='center', fontdict={'family':'serif','size':20,\
    'weight':'normal','color':'black'})
plt.text(65, 500, r'$\mu=100,\ \sigma=10$');
```

```python
# Modifying text attributes:
plt.hist(x, bins=50, facecolor='r', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.title('Histogram', loc='center', fontdict={'family':'serif','size':20,\
    'weight':'normal','color':'black'})
plt.text(65, 500, r'$\mu=100,\ \sigma=10$', fontdict={'size':12,\
    'weight':'bold', 'color':'darkred'});
```

```python
# Modifying facecolor='g'
plt.hist(x, bins=50, facecolor='g', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.title('Histogram', loc='center', fontdict={'family':'serif','size':20,\
    'weight':'normal','color':'black'})
plt.text(65, 500, r'$\mu=100,\ \sigma=10$', fontdict={'size':12, \
    'weight':'bold', 'color':'darkgreen'});
```

```python
# Modifying facecolor='b'
plt.hist(x, bins=50, facecolor='b', alpha=0.6)
plt.xlabel('x values',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.title('Histogram', loc='center', fontdict={'family':'serif','size':20,\
    'weight':'normal','color':'black'})
plt.text(65, 500, r'$\mu=100,\ \sigma=10$', fontdict={'size':12, \
    'weight':'bold', 'color':'darkblue'});
```

### Adding annotations with arrows

The uses of `text()` will place text at an arbitrary position on the Axes. A common use case of text is to annotate some feature of the plot, and the `annotate()` method provides helper functionality to make annotations easy.

Let's use a different example:

```python
x = np.arange(0.0, 6.0, 0.01)
y = np.sin(2*np.pi*x)
line, = plt.plot(x, y)

# Adding annotations
plt.annotate('local max', xy=(2.3, 1), xytext=(3, 1.5),arrowprops=dict(facecolor='darkgreen'),)
plt.annotate('local min', xy=(2.8, -1), xytext=(3.5, -1.5),arrowprops=dict(facecolor='darkred'),)

plt.ylim(-2, 2)
plt.show()
```
