# Categorical Plot in Seaborn

## Objectives

- To explore the capabilities of Seaborn for creating visually appealing categorical plots.
- To demonstrate various categorical plots, including strip plots, swarm, box, violin, and bar plots.
- To highlight the customizability of categorical plots regarding data representation and aesthetic adjustments.

## Background

Seaborn extends Matplotlib's functionality with support for categorical data visualization, simplifying the creation of complex visualizations. The library provides specialized plot types explicitly designed for categorical variables, making interpreting the data's structure and distributions easier.

## Datasets Used

The tips dataset loaded from Seaborn is used to demonstrate categorical plotting. 

## Loading Tips Dataset

```python
import numpy as np
import seaborn as sns
```

Load an example dataset:

```python
tips = sns.load_dataset("tips")
tips.head()
```

Catplot is simplifies plotting that involves categorical variables. These plots offer different levels of granularity.

## Categorical Scatterplots

- `stripplot()` – with `kind="strip"`
- `swarmplot()` – with `kind="swarm"`

### Strip Plot

A `strip` is basically a scatter plot where the x axis represents a categorical variable. 

Typical uses of a strip plot involves applying a small random jitter value to each data point such that the separation between points becomes clearer.

strip plot

```python
sns.catplot(data=tips, x="day", y="total_bill");
```

`kind='strip'` is the default value 

```python
sns.catplot(data=tips, x="day", y="total_bill", kind="strip");
```

For changing the order of categories

```python
category_order = ['Sat', 'Sun', 'Thur', 'Fri']
```

Changing the order of categories

```python
sns.catplot(data=tips, x="day", y="total_bill", 
            order=category_order);
```

Adding a title

```python
sns.catplot(data=tips, x="day", y="total_bill", 
            order=category_order).set(title="Total Bill by Day");
```

The jitter parameter controls the magnitude of jitter or disables it altogether.

Specifying `jitter=False` to remove the jitter

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='strip', 
            order=category_order, jitter=False).set(title="Total Bill by Day");
```

### Swarm Plot

kind='swarm'

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='swarm', 
            order=category_order).set(title="Total Bill by Day");
```

Changing the role of `x` and `y`

```python
sns.catplot(data=tips, y="day", x="total_bill", kind='swarm', 
            order=category_order).set(title="Total Bill by Day");
```

## Categorical Distribution Plots

- `boxplot()` – with `kind="box"`
- `boxenplot()` – with `kind="boxen"`
- `violinplot()` – with `kind="violin"`

### Box Plot

kind='box'

```python
# kind='box'
sns.catplot(data=tips, x="day", y="total_bill", kind='box', 
            order=category_order).set(title="Total Bill by Day");
```

Black dots represent outliers. 

### Boxen Plot

Boxenplot graph is an enhanced boxplot, because it shows a large number of quantiles. It is similar to box plot in graphing a nonparametric representation of a data distribution. 

kind='boxen'

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='boxen', 
            order=category_order).set(title="Total Bill by Day");
```

`showfliers=False` suppress the plotting of outliers

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='boxen', 
            order=category_order, showfliers=False).set(title="Total Bill by Day");
```

## Violin Plot

Violin plots are similar to box plots, except they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.

kind='violin'

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin', 
            order=category_order).set(title="Total Bill by Day");
```

You can change the representation of the datapoints in the violin interior:
- `inner='box'` will draw a miniature boxplot (default option)
- `inner='quartiles'` will draw the quartiles of the distribution 
- `inner='point'` will show each underlying datapoint
- `inner='stick'` will show each underlying datapoint 
- `inner=None`: will draw unadorned violins

kind='violin', inner='quartiles' 

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin', 
            inner='quartiles', order=category_order).set(title="Total Bill by Day");
```

kind='violin', inner='point' 

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin', 
            inner='point', order=category_order).set(title="Total Bill by Day");
```

kind='violin', inner='stick' 

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin', 
            inner='stick', order=category_order).set(title="Total Bill by Day");
```

kind='violin', inner=None

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin', 
            inner=None, order=category_order).set(title="Total Bill by Day");
```

## Categorical Estimate Plots with catplot

- `countplot()` - with `kind="count"`
- `pointplot()` – with `kind="point"`
- `barplot()` – with `kind="bar"`

### Count Plot

A count plot displays the number of observations in each category of your variable. 

```python
sns.catplot(data=tips, x="day", kind='count',
            order=category_order).set(title='Number of Days');
```

Horizontal bars (change x for y)

```python
sns.catplot(data=tips, y="day", kind='count',
            order=category_order).set(title='Number of Days');
```

### Point Plot

A `point` plot shows mean estimate and uncertainty of the estimate with a point and error bar for each categorical variable. 

It is a great way to visualize the interaction between different variables.

kind='point'

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='point', 
            order=category_order);
```

With the parameter `estimator` you can change the statistical function to apply inside each bin. 

Default `estimator=np.mean`

```python
# estimator=np.mean produces the same chart
sns.catplot(data=tips, x="day", y="total_bill", kind='point', 
            estimator=np.mean, order=category_order).set(title="Mean of Total Bill by Day");
```

estimator=np.median

```python
# estimator=np.median
sns.catplot(data=tips, x="day", y="total_bill", kind='point', 
            estimator=np.median, order=category_order).set(title="Median of Total Bill by Day");
```

estimator=np.median, changing x and y

```python
sns.catplot(data=tips, y="day", x="total_bill", kind='point', 
            estimator=np.median, order=category_order).set(title="Median of Total Bill by Day");
```

```python
sns.catplot(data=tips, y="day", x="total_bill", kind='point', 
            estimator=np.median, linestyle='--', 
            order=category_order).set(title="Median of Total Bill by Day");
```

Without connecting the points

```python
sns.catplot(data=tips, y="day", x="total_bill", kind='point', 
            estimator=np.median, linestyle='none', 
            order=category_order).set(title="Median of Total Bill by Day");
```

### Bar Plot

Another popular choice for plotting categorical data is a bar plot.

The height of each bar represents the `mean` value in each category. 

kind='bar'

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='bar', 
            order=category_order).set(title="Mean of Total Bill by Day");
```

Black lines at the top of each bar represent 95% confidence intervals for the mean. You can turn off confidence intervals setting the ci parameter to None.

kind='bar' without confidence interval

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='bar', 
            errorbar=None, order=category_order).set(title="Mean of Total Bill by Day");
```

Calculating the median of the total bill by day

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='bar', 
            estimator=np.median, order=category_order).set(title="Median of Total Bill by Day");
```

Calculating the sum of the total bill by day

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='bar', 
            estimator=np.sum, order=category_order).set(title="Sum of Total Bill by Day");
```

## Bivariate Graphs

Adding the `hue` parameter

```python
sns.catplot(data=tips, x="day", y="total_bill",
            order=category_order, hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='swarm', 
            order=category_order, hue='smoker');
```

Adding the `col` parameter

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='box', 
            order=category_order, hue='smoker', col='time');
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='boxen',  
            order=category_order, hue='smoker', col='time');
```

Adding the `row` parameter

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin', 
            order=category_order, hue='smoker', col='time', row='sex');
```

```python
sns.catplot(data=tips, x="day", y="total_bill", kind='violin',  
            order=category_order, hue='smoker', inner='stick', split=True);
```

Changing the role of `x` and `y`

```python
sns.catplot(data=tips, y="day", x="total_bill", kind='violin',  
            order=category_order, hue='smoker', inner='stick', split=True);
```

## Conclusions

Key Takeaways:
- Seaborn supports a range of plot types specifically for categorical data, such as strip plots for detailed scatter-style visualizations and box, boxen, and violin plots for exploring data distribution and outliers.
- The library offers significant customization options, including category ordering, jitter adjustment, outlier management, and data representation choices within plots, enhancing flexibility in data presentation.
- Its seamless integration with pandas DataFrames allows direct references to DataFrame columns in plot functions, simplifying data visualization.
- Seaborn not only displays data but also integrates statistical insights directly into the visualizations, such as mean and median calculations through point plots and bar plots, providing a rich context for analysis.
