# Categorical Plot in Seaborn

```python
import seaborn as sns
```

```python
# Apply the default theme
sns.set_theme()
```

The `set_theme()` will affect how all matplotlib plots look, even if you don’t make them with seaborn.

```python
# Load an example dataset
tips = sns.load_dataset("tips")
tips.head()
```

Catplot is simplifies plotting that involves categorical variables. These plots offer different levels of granularity.

Categorical scatterplots with catplot
- stripplot() – with kind=”strip”
- swarmplot() – with kind=”swarm”

```python
sns.catplot(data=tips, x="day", y="total_bill");
```

Swarm plot (`kind='swarm'`) is a scatter plot that adjusts the positions of the points along the categorical axis so that they don’t overlap.

```python
# kind='swarm'
sns.catplot(data=tips, x="day", y="total_bill",
            kind='swarm');
```

Categorical distribution plots with catplot
- boxplot() – with kind=”box”
- boxenplot() – with kind=”boxen”
- violinplot() – with kind=”violin”

```python
# kind='box'
sns.catplot(data=tips, x="day", y="total_bill",
            kind='box');
```

```python
# kind='boxen'
sns.catplot(data=tips, x="day", y="total_bill",
            kind='boxen');
```

```python
# kind='violin'
sns.catplot(data=tips, x="day", y="total_bill",
            kind='violin');
```

Categorical estimate plots with catplot
- pointplot() – with kind=”point”
- barplot() – with kind=”bar”

```python
# kind='point'
sns.catplot(data=tips, x="day", y="total_bill",
            kind='point');
```

```python
# kind='bar'
sns.catplot(data=tips, x="day", y="total_bill",
            kind='bar');
```

Including the `hue` parameter.

```python
sns.catplot(data=tips, x="day", y="total_bill",
            hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='swarm', hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='box', hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='boxen', hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='violin', hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='violin', hue='smoker', split=True);
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='point', hue='smoker');
```

```python
sns.catplot(data=tips, x="day", y="total_bill",
            kind='bar', hue='smoker');
```
